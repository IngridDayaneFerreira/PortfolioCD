```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts

dados_covid = pd.read_csv("https://raw.githubusercontent.com/gcpeixoto/ICD/main/database/covid19-weekly-trends-in-europe.csv")
dados_covid.head()
```

```python
paises_UE = pd.Series({'UE': ['Germany', 'Austria', 'Belgium', 'Bulgaria', 'Cyprus', 'Croatia', 'Denmark', 'Slovakia',
                             'Slovenia', 'Spain','Estonia', 'Finland', 'France', 'Greece', 'Netherlands', 'Hungary',
                             'Ireland', 'Italy', 'Latvia', 'Lithuania','Luxembourg', 'Malta', 'Poland', 'Portugal',
                             'Czech Republic', 'Romania', 'Sweden']})

dados_covid_1 = dados_covid.drop(['Cases in the preceding 7 days', 'Weekly Case % Change', 'Cases in the last 7 days/1M pop',
                 'Deaths in the last 7 days', 'Deaths in the preceeding 7 days', 'Weekly Death % Change',
                 'Deaths in the last 7 days/1M pop', 'Population'], axis=1)
dados_covid_1
```

```python
UE = dados_covid_1

for i in range(UE['Country, Other'].count()):
    if UE.loc[i]['Country, Other'] not in paises_UE[0]:
        UE = UE.drop([i], axis=0)
        
UE
```

```python
non_UE = dados_covid_1

for i in range(non_UE['Country, Other'].count()):
    if non_UE.loc[i]['Country, Other'] in paises_UE[0]:
        non_UE = non_UE.drop([i], axis=0)
        
non_UE
```

```python
def p(x):
    return 3*( x.mean() - x.median() )*x.std()

series_UE = pd.Series(UE['Cases in the last 7 days'])
person_UE = p(series_UE)

series_non_UE = pd.Series(non_UE['Cases in the last 7 days'])
person_non_UE = p(series_non_UE)

person_UE, person_non_UE
```
