```python
def cov(df,nome_coluna1,nome_coluna2):
    s1,s2 = df[nome_coluna1],df[nome_coluna2]
    return np.dot( s1 - s1.mean(), s2 - s2.mean() )/(len(s1)-1)

#Covariâncias
sequencia_covariancias = {'covariancia PTS x J' : cov(brasileirao,'PTS','J'), 'covariancia PTS x V' : cov(brasileirao,'PTS','V'), 
                          'covariancia PTS x E' : cov(brasileirao,'PTS','E'), 'covariancia PTS x D' : cov(brasileirao,'PTS','D'), 
                          'covariancia PTS x GP' : cov(brasileirao,'PTS','GP'), 'covariancia PTS x GC' : cov(brasileirao,'PTS','GC'), 
                          'covariancia PTS x SG' : cov(brasileirao,'PTS','SG'), 'covariancia PTS x CA' : cov(brasileirao,'PTS','CA'), 
                          'covariancia PTS x CV' : cov(brasileirao,'PTS','CV')}

sequencia_covariancias
```

```python
sorted(sequencia_covariancias.items(), key=lambda x: x[1], reverse=True)
```
