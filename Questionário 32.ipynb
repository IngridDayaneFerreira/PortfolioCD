```python
#Questão 2

import numpy as np
import pandas as pd

serie_estado = pd.Series({'VW Gol': 'Bahia', 'BMW Z4': 'Ceará', 'Chery QQ': 'Sergipe'})
serie_distancia = pd.Series({'VW Gol': 848, 'BMW Z4': 728, 'Chery QQ': 640})
serie_duracao = pd.Series({'VW Gol': 7.5, 'BMW Z4': 10, 'Chery QQ': 9.5})

dicionario_series = {'Estado': serie_estado, 'Distancia de JPA (km)': serie_distancia, 'Duracao (h)': serie_duracao}

df = pd.DataFrame(dicionario_series)

df
```

```python
vm_gol = (df.loc['VW Gol']['Distancia de JPA (km)']/df.loc['VW Gol']['Duracao (h)'])/3.6
vm_bmw = (df.loc['BMW Z4']['Distancia de JPA (km)']/df.loc['BMW Z4']['Duracao (h)'])/3.6
vm_qq = (df.loc['Chery QQ']['Distancia de JPA (km)']/df.loc['Chery QQ']['Duracao (h)'])/3.6

serie_vel_media = ({'VW Gol': vm_gol, 'BMW Z4': vm_bmw, 'Chery QQ': vm_qq})
dicionario_series = {'Estado': serie_estado, 'Distancia de JPA (km)': serie_distancia, 'Duracao (h)': serie_duracao, 'Vel. Inicial (m/s)': serie_vel_media}
df = pd.DataFrame(dicionario_series)

vin_gol = ((df.loc['VW Gol']['Distancia de JPA (km)']/df.loc['VW Gol']['Duracao (h)'])/3.6) + (df.loc['VW Gol']['Vel. Inicial (m/s)'] * 0.12)
vin_bmw = (df.loc['BMW Z4']['Distancia de JPA (km)']/df.loc['BMW Z4']['Duracao (h)'])/3.6 + (df.loc['BMW Z4']['Vel. Inicial (m/s)'] * 0.12)
vin_qq = (df.loc['Chery QQ']['Distancia de JPA (km)']/df.loc['Chery QQ']['Duracao (h)'])/3.6 + (df.loc['Chery QQ']['Vel. Inicial (m/s)'] * 0.12)

serie_vel_media_pos_aceleracao = ({'VW Gol':  vin_gol, 'BMW Z4': vin_bmw, 'Chery QQ': vin_qq})
dicionario_series = {'Estado': serie_estado, 'Distancia de JPA (km)': serie_distancia, 'Duracao (h)': serie_duracao, 'Vel. Inicial (m/s)': serie_vel_media, 'Vel. Final (m/s)': serie_vel_media_pos_aceleracao}
df = pd.DataFrame(dicionario_series)

ac_gol = (df.loc['VW Gol']['Vel. Final (m/s)'] - df.loc['VW Gol']['Vel. Inicial (m/s)']) / 120
ac_bmw = (df.loc['BMW Z4']['Vel. Final (m/s)'] - df.loc['BMW Z4']['Vel. Inicial (m/s)']) / 120
ac_qq = (df.loc['Chery QQ']['Vel. Final (m/s)'] - df.loc['Chery QQ']['Vel. Inicial (m/s)']) / 120

serie_aceleracao = ({'VW Gol': ac_gol, 'BMW Z4': ac_bmw, 'Chery QQ': ac_qq})
dicionario_series = {'Estado': serie_estado, 'Distancia de JPA (km)': serie_distancia, 'Duracao (h)': serie_duracao, 'Vel. Inicial (m/s)': serie_vel_media, 'Vel. Final (m/s)': serie_vel_media_pos_aceleracao, 'Aceleração (m/s2)': serie_aceleracao}
df = pd.DataFrame(dicionario_series)

df
```

```python
a = df.loc['VW Gol']['Vel. Inicial (m/s)']
b = df.loc['BMW Z4']['Vel. Final (m/s)']
c = df.loc['Chery QQ']['Aceleração (m/s2)']

print('(%.2f, %.2f, %.2f)' %(a,b,c))
```
