```python
#transformando em int
arquivoEnem[['Média da nota da prova de Ciências da Natureza','Média da nota da prova de Ciências Humanas','Média da nota da prova de Linguagens e Códigos','Média da nota da prova de Matemática','Média da nota da prova de redação']] = arquivoEnem[['Média da nota da prova de Ciências da Natureza','Média da nota da prova de Ciências Humanas','Média da nota da prova de Linguagens e Códigos','Média da nota da prova de Matemática','Média da nota da prova de redação']].astype('int64')
arquivoEnem
```

```python
enemCH = arquivoEnem[['Estado','Média da nota da prova de Ciências Humanas']]
enemCH.sort_values(['Média da nota da prova de Ciências Humanas'])
```

```python
amp_total = np.max(enemCH['Média da nota da prova de Ciências Humanas']) - np.min(enemCH['Média da nota da prova de Ciências Humanas'])
amp_total
```

```python
n = enemCH['Média da nota da prova de Ciências Humanas'].count()
n

k = 1 + (3.22 * np.log10(n))

k
```

```python
amp_classes = amp_total / k
h = round(amp_classes)
h
```

```python
frequencias = []
menor = enemCH['Média da nota da prova de Ciências Humanas'].min()

valor = menor
while (valor < enemCH['Média da nota da prova de Ciências Humanas'].max()):
    frequencias.append(valor)
    valor += h
frequencias
```

```python
f = pd.cut(enemCH['Média da nota da prova de Ciências Humanas'], bins=frequencias, right=False)
f
```

```python
#Questão 2

arqEnemNovo = pd.read_excel('enem2019.xlsx')
for i in range(arqEnemNovo['COD UF'].count()):
    if ((arqEnemNovo.loc[i]['COD UF'] // 10) != 2):
        arqEnemNovo = arqEnemNovo.drop([i], axis=0)
arqEnemNovo
```
