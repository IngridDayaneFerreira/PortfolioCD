```python
#Quest達o 1

import numpy as np
import pandas as pd

img = np.load('SyntEyeKTC-106_a.npz')['arr_0']

col = pd.read_csv('SyntEyeKTC-106_a.csv')

imed = int(np.round(((col["y"].max() - col["y"].min()) / 2), 0))

imed
```

```python
#Quest達o 2

import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import lineplot

col_0 = col[col['x'] == imed]

fig1, ax = plt.subplots()
ax = sns.lineplot(x = col_0["y"] ,y = col_0["r"], data = col_0)
ax.set_title("(a)")
ax.set_ylabel("Vermelho")
ax.set_xlabel("Pixel sobre a linha central")

fig2, bx = plt.subplots()
bx = sns.lineplot(x = col_0["y"] ,y = col_0["b"], data = col_0)
bx.set_title("(b)")
bx.set_ylabel("Azul")
bx.set_xlabel("Pixel sobre a linha central")

fig3, cx = plt.subplots()
cx = sns.lineplot(x = col_0["y"] ,y = col_0["g"], data = col_0)
cx.set_title("(c)")
cx.set_ylabel("Verde")
cx.set_xlabel("Pixel sobre a linha central")

print("(a): vermelho, (b): azul, (c): verde.")
```

```python
#Quest達o 3

colf = col.loc[(col.r >= 50) & (col.r <= 200) & (col.g >= 50) & (col.g <= 200) & (col.b >= 50) & (col.b <= 200)]

print("(a): vermelho x verde, (b): vermelho x azul, (c): azul x verde")

dma = sns.jointplot(x = "r", y = "g", data = colf, height = 5, color = "blue")
title_a = plt.suptitle("(a): Vermelho x Verde", y = 1)

dmb = sns.jointplot(x = "r", y = "b", data = colf, height = 5, color = "orange")
title_b = plt.suptitle("(b): Vermelho x Azul", y = 1)

dmc = sns.jointplot(x = "b", y = "g", data = colf, height = 5, color = "purple")
title_c = plt.suptitle("(c): Azul x Verde", y = 1)
```
